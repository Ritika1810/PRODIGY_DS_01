```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
url = "https://raw.githubusercontent.com/kindo-tk/PRODIGY_DS_01/main/worldpopulationdata.csv"
df = pd.read_csv(url)
print(df.columns)

# Create histogram
plt.figure(figsize=(10, 5))
plt.hist(df['2021'], bins=20, color='skyblue', edgecolor='black')
plt.title('Population Distribution in 2021')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('population_histogram_2021.png')
plt.show()
```
